#### In this notebook, I am looking to find out how many of each type (customer or subscriber) of Citi Bike users use Citi Bikes each day. CitiBike's data is available to the public. I am only using one day's worth of data.

```python
import csv
```

```python
source_file = open("202009CitibikeTripdataExample.csv","r")
```

#### In this cell I am taking a look at the data, I am looking through all the column names to see which column the correct, needed information is found in.

```python
citibike_reader = csv.DictReader(source_file)

print(citibike_reader.fieldnames)
```

#### Since at the end of the notebook, we want a count of how many customers vs. subscribers rode citibikes, we are setting the count of "subscribers", "customers", (and "other" just in case) to 0.

```python
subscriber_count = 0
customer_count = 0
other_user_count = 0
```

#### This is the main part, we set an iterative function that adds 1 to the "subscriber" count for every that has a subscriber, for every row with a customer 1 is added to the "customer" count, if the column is empty or there is something that is neither customer nor subscriber, 1 gets added to the "other" count.

```python
for a_row in citibike_reader:

    if a_row["usertype"] == "Subscriber":
          subscriber_count = subscriber_count +1
    elif a_row["usertype"] == "Customer":

        customer_count = customer_count + 1

    else:
        other_user_count = other_user_count + 1
```

#### In this cell we print the results:
#### Most of the users of citibikes are subscribers.

```python
print("Number of subscribers:")
print(subscriber_count)
print("Number of customers:")
print(customer_count)
print("Number of 'other' users:")
print(other_user_count)
```
